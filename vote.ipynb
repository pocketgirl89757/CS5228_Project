```python
folder = 'results/vote1'

import glob
import pandas as pd

files = glob.glob(folder + "/*.csv")

full_df = None

for filename in files:
    if full_df is None:
        full_df = pd.read_csv(filename)
        full_df = full_df.drop(columns=['Id'])
        full_df = full_df.rename(columns={"ChargeOff": filename})
    else:
        df = pd.read_csv(filename)
        df = df.drop(columns=['Id'])
        df = df.rename(columns={"ChargeOff": filename})
        full_df = pd.concat([full_df, df], axis=1)

full_df['majority'] = full_df.mode(axis=1)[0]


pd.DataFrame(full_df['majority']).to_csv(folder +'.csv',header=['ChargeOff'],index_label="Id")

full_df.head(30)
```

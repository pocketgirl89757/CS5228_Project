```python
df = get_data(type='train',get_dummy=False,dropna=True, values_only=False)

plt.figure(figsize=(20,5))
sb.countplot(x='State', data=df,order = df['State'].value_counts().index,hue='ChargeOff');

plt.figure(figsize=(20,5))
sb.countplot(x='BankState', data=df,order = df['BankState'].value_counts().index,hue='ChargeOff' );

plt.figure(figsize=(25,5))
sb.countplot(x='ApprovalFY', data=df.sort_values(by='ApprovalFY'),hue='ChargeOff');

plt.figure(figsize=(5,5))
sb.countplot(x='NewExist', data=df,order = df['NewExist'].value_counts().index,hue='ChargeOff' );

plt.figure(figsize=(5,5))
sb.countplot(x='UrbanRural', data=df,order = df['UrbanRural'].value_counts().index,hue='ChargeOff' );

plt.figure(figsize=(5,5))
sb.countplot(x='RevLineCr', data=df,order = df['RevLineCr'].value_counts().index,hue='ChargeOff' );

plt.figure(figsize=(5,5))
sb.countplot(x='LowDoc', data=df,order = df['LowDoc'].value_counts().index,hue='ChargeOff' );

columns_for_plot=['Term','NoEmp','DisbursementGross','GrAppv','SBA_Appv']
df.hist(bins=50,column = columns_for_plot,figsize=[20,15]);

# outliner for NoEmp
```

```python
plt.figure(figsize=(5,5))
plt.title('NoEmp')
plt.boxplot(df['NoEmp'])

plt.figure(figsize=(5,5))
plt.title('Term')
plt.boxplot(df['Term'])

plt.figure(figsize=(5,5))
plt.title('DisbursementGross')
plt.boxplot(df['DisbursementGross'])

plt.figure(figsize=(5,5))
plt.title('GrAppv')
plt.boxplot(df['GrAppv'])

plt.figure(figsize=(5,5))
plt.title('SBA_Appv')
plt.boxplot(df['SBA_Appv'])

```
